## An end to end Scikit Learn workflow

```python
#1.Get the dta ready
import pandas as pd
import numpy as np
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#Creating features matrix (X)
X=heart_disease.drop("target",axis=1)

#Creating Labels (Y)
Y=heart_disease["target"]
```

```python
#2. Choose the right model and hyper-parametres.
from sklearn.ensemble import RandomForestClassifier
#randomforestclassifier is a classification machine learning model.
clf= RandomForestClassifier()

#We'll keep the default hyper parametres.
clf.get_params()
```

```python
#3.Fit the model to the training data
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
clf.fit(X_train,Y_train);
```

```python
#making prediction
y_preds=clf.predict(X_test)
y_preds
```

```python
Y_test
```

```python
#4. Evaluate the model
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(Y_test,y_preds))
```

```python
confusion_matrix(Y_test,y_preds)
```

```python
accuracy_score(Y_test,y_preds)
```

```python
#5.Improve a model
#Try diff amt of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators")
    clf=RandomForestClassifier(n_estimators=i)
    clf.fit(X_train,Y_train)
    print(f"Model accuracy on test set:{clf.score(X_test,Y_test)}")    
```

```python
#6. Saving and reloading a model.
import pickle
pickle.dump(clf, open("random_forest_model_1.pk1","wb"))
#wb for binary!
```

```python
loaded_model=pickle.load(open("random_forest_model_1.pk1","rb"))
```

```python
loaded_model.score(X_test,Y_test)
```
