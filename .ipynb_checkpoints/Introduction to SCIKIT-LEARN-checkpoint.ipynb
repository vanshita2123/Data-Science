#### 1.1 Making sure its all numerical.

```python
car_sales=pd.read_csv("Data-and-images/car-sales-extended.csv")
```

```python
car_sales.head()
```

```python
len(car_sales)
```

```python
car_sales.dtypes
```

```python
#Splitting the data into x and y
x=car_sales.drop("Price",axis=1)
y=car_sales["Price"]

#Splitting into training and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

#Build ML model


from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)

--This code will generate an output--

```python
#Error as expected! ML model can't be trained on strings.
#Turning the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Doors is a categorical feature too!
categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],remainder="passthrough")
transformed_x=transformer.fit_transform(x)
```

OneHotEncoder converts everything into numbers!  
remainder=passthrough is used so that all other fields other than categorical passes as it is.

```python
transformed_x
```

```python
pd.DataFrame(transformed_x)
```

```python
#Another way to convert into numbers.
dummies=pd.get_dummies(car_sales[["Make","Colour","Doors"]])
dummies
```

```python
#Lets refit the model.
np.random.seed(42)
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
