```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("Data-and-images/sample-plot.png")
```

## Making figures with NumPy Arrays

```python
import numpy as np
```

```python
#Creating data
x=np.linspace(0,10,100)
x
```

```python
#Creating line plot (default)
fig, ax= plt.subplots()
ax.plot(x,x**2);
```

```python
#Creating scatter plot
fig,ax=plt.subplots()
ax.scatter(x,x**x)
```

```python
a=[1,4,2,4,7,2,9]
b=[2,5,3,7,4,9,10]
fig,ax=plt.subplots()
ax.scatter(a,b)
```

```python
#Creating bar plot
fig, ax=plt.subplots()
ax.bar(a,b)
```

```python
#Creating bar plot using dictionary
fig, ax= plt.subplots()
Chinese_Pricelist= {"Noodles":110,"Fried_rice":90,"Manchurian":200}
ax.bar(Chinese_Pricelist.keys(),Chinese_Pricelist.values())
ax.set(title="Best-Junk",xlabel="Items",ylabel="Price");
```

```python
#Creating histogram 
x= np.random.randint(100, size=100)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
x=np.random.randn(10000)
#randn creates a normal distribution.
fig,ax=plt.subplots()
ax.hist(x);
```

```python
#Creating subplots
#Option1
fig,(ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(10,10))
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax1.plot(a,b);
ax2.scatter(x,x**2);
ax3.hist(np.random.random(10));
ax4.bar(a,b);
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.head()
```

```python
#NON IDEAL PYPLOT GRAPHS:
#Non ideal because it uses the same scale for all items.
heart_disease.plot.hist(subplots=True,figsize=(10,20));
```

```python
#Using Object orientred ideal way:
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
#oo method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', 
             x="age", 
             y="chol", 
             c='target', 
             ax=ax);
ax.set_xlim([45, 100]);
```

```python
#OO method:
fig, ax = plt.subplots(figsize=(10, 6))
Figuree= ax.scatter(over_50["age"], 
                     over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#Adding a legend
ax.legend(*Figuree.legend_elements(), title="Target");

#Adding a mean horizontal line.
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
# Subplot of chol,thalach,age.
plt.style.use('default')
fig, (ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True);

#Adding data
Figure1=ax1.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
Figure2=ax2.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"]);

#Customising data
ax1.set(title="Heart disease and cholestrol",xlabel="Age",ylabel="Cholestrol")
ax2.set(title="Heart disease and max heart rate", xlabel="Age", ylabel="Thalach")

#Adding legend
ax1.legend(*Figure1.legend_elements(),title="Target")
ax2.legend(*Figure2.legend_elements(),title="Target")

#Adding mean line
ax1.axhline(over_50["chol"].mean(),linestyle="--");
ax2.axhline(over_50["thalach"].mean(),linestyle="--");

#Adding a title to fig
fig.suptitle("Heart disease analysis", fontsize=20, fontweight="bold");
```
